1. грядка, есть/кушать, зелёный

```python
# !pip install git+https://github.com/lopuhin/python-adagram.git
import adagram
from nltk.tokenize import RegexpTokenizer
from pymorphy2 import MorphAnalyzer
import numpy as np
import re
from pymystem3 import Mystem
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.pipeline import Pipeline

token = RegexpTokenizer('\w+')
morph = MorphAnalyzer()

def normalize_pm(text):
    words = [morph.parse(word)[0].normal_form for word in tokenize(text) if word]
    return words

def tokenize(text):
    return token.tokenize(text)
```

```python
vm = adagram.VectorModel.load('all.a010.p10.d300.w5.m100.nonorm.slim.joblib')
```

```python
results_1 = []
for sentence in sentences_1:
    means = vm.disambiguate(lexeme, normalize_pm(sentence))
    results_1.append(np.argmax(means))
    print(results_1[-1], sentence)
    print(vm.sense_neighbors('лук', results_1[-1]))
    print()
```

```python
results_2 = []
for sentence in sentences_2:
    means = vm.disambiguate(lexeme, normalize_pm(sentence))
    results_2.append(np.argmax(means))
    print(results_2[-1], sentence)
    print(vm.sense_neighbors('лук', results_2[-1]))
    print()
```

Второе значение распознается гораздо лучше, чем первое. Не очень понятно, в чём принципиальное отличие 1 и 4 значений, так же, как и между 0 и 2-м значением. Отдельное 3-е значение (Лука, мужское имя евангелиста) совсем ошибочно присваивается в предложении "Закусили. ― Я в этом году первый раз зеленый лук ем, ― сказала Александра."

4. Соотнесём примеры с конкретными толкованиями из wiki_ru_wordnet-а с помощью tf-idf

```python
from wiki_ru_wordnet import WikiWordnet
```

```python
wikiwordnet = WikiWordnet()
```

```python
synsets = wikiwordnet.get_synsets(lexeme)
```

```python
original_definitions = []
for synset in synsets:
    for w in synset.get_words():
        if w.lemma() == lexeme:
            original_definitions.append(w.definition())
```

```python
def preprocess_text(text):
    text = [word for word in tokenize(text) if not re.search('[a-zA-Z0-9]', word)]
    text = [w.strip().lower() for w in mystem.lemmatize(' '.join(text)) if w.strip() and w != lexeme and w != 'пример']
    return text
```

```python
mystem = Mystem()

definitions = []
for i, original_definition in enumerate(original_definitions):
    definitions.append(preprocess_text(original_definition))
    print(definitions[i], '\n')
```

```python
tokenized_sentences_1 = []
for sentence_1 in sentences_1:
    tokenized_sentences_1.append(preprocess_text(sentence_1))
```

```python
tokenized_sentences_2 = []
for sentence_2 in sentences_2:
    tokenized_sentences_2.append(preprocess_text(sentence_2))
```

```python
corpus = [
    *[' '.join(sent) for sent in tokenized_sentences_1],
    *[' '.join(sent) for sent in tokenized_sentences_2],
    *[' '.join(definition) for definition in definitions]
]
```

```python
pipe = Pipeline([
    ('count', CountVectorizer()),
    ('tfidf', TfidfTransformer())
])
vectors = pipe.fit_transform(corpus).toarray()
first_dataset = vectors[:8]
second_dataset = vectors[8:16]
definitions_dataset = vectors[16:]
```

```python
def print_closest_definitions(dataset):
    results = []
    for i, vector in enumerate(dataset):
        cosines = []
        for def_vector in definitions_dataset:
            cosines.append(cosine_similarity(vector.reshape(1, -1), def_vector.reshape(1, -1)))
        result = np.argmax(cosines)
        results.append(result)
        print(f'{i}-й пример: {result}')
    print()
    for i in set(results):
        print(f'{i}-е определение: {original_definitions[i]}\n')
    return results
```

```python
closest_definitions_1 = print_closest_definitions(first_dataset)
```

```python
closest_definitions_2 = print_closest_definitions(second_dataset)
```

На первом датасете ошибок нет, поскольку МАС не различает 1-е и 2-е определения, однако на втором датасете tf-idf справляется гораздо хуже, видимо, потому что отсутствует слово "стрелять" в 5-м определении.

```python
map_ = {
    0: 2,
    1: 5,
    2: 2,
    3: -1,
    4: 5
}
```

```python
def count_accuracy(lst_1, lst_2):
    tp = 0
    for a, b in zip(lst_1, lst_2):
        if map_[a] == b:
            tp += 1
    return tp / min(len(lst_1), len(lst_2))
```

```python
print(f'accuracy (1): {count_accuracy(results_1, closest_definitions_1)}')
```

```python
print(f'accuracy (2): {count_accuracy(results_2, closest_definitions_2)}')
```

```python
print(f'accuracy (1+2): {count_accuracy([*results_1, *results_2], [*closest_definitions_1, *closest_definitions_2])}')
```
